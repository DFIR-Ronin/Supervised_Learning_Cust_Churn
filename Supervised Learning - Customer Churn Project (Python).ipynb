```python
# data preprocessing

data.isna().sum() # so we dont have non data
```

```python
data.drop(columns="customerID",inplace=True)# 
data
```

```python
for col in data.columns:
  print(col)
  print(len(data[col].unique())) # show number of unique data in each column
```

```python
for i in range(len(data["TotalCharges"])): # covert str to float in TotalCharges
  if data["TotalCharges"][i]==" ":
    data["TotalCharges"][i]=0.0
  else:


    # print(i)
    # print(data["TotalCharges"][i])
    num=float(data["TotalCharges"][i])
    data["TotalCharges"][i]=0.0
    data["TotalCharges"][i]=num
```

```python
for col in data.columns:  # outliers in numeric dataset  ====> they dont have 
  if data[col].dtype=="int" or data[col].dtype=="float":
    print(col)
    upper_range=data[col].mean()+ 3*data[col].std()
    lower_range=data[col].mean()- 3*data[col].std()

    indexs=data[(data[col]>upper_range) | (data[col]<lower_range) ].index
    
    print(indexs)
    data=data.drop(indexs)
```

```python
upper_range=data["TotalCharges"].mean()+ 3*data["TotalCharges"].std() # delete outliers from TotalCharges
lower_range=data["TotalCharges"].mean()- 3*data["TotalCharges"].std()
indexs=data[(data["TotalCharges"]>upper_range) | (data["TotalCharges"]<lower_range) ].index

data=data.drop(indexs)
```

```python
data["tenure"]=pd.cut(data["tenure"],3,labels=["low","medium","large"]) # discretization tenure to 3 section 
```

```python
data["MonthlyCharges"]=pd.cut(data["MonthlyCharges"],4,labels=["low","medium","large","very high"])# discretization MonthlyCharges to 4 section 
```

```python
data["TotalCharges"]=pd.cut(data["TotalCharges"],4,labels=["low","medium","large","very high"])# discretization TotalCharges to 4 section 
```

```python
dummier_fitures=pd.get_dummies(data.iloc[:,:-1]) # use one_hot_encoding in categorical dataset
dummier_fitures
```

```python
finally_data=pd.concat([dummier_fitures,data["Churn"]],axis=1)#concat dummier_fitures and data["Churn"]
```

```python
finally_data
```

```python
from sklearn.preprocessing import LabelEncoder # use label encoder to convert churn to muneric
finally_data["Churn"]=LabelEncoder().fit_transform(finally_data["Churn"])
```

```python
finally_data
```

```python
x=finally_data.iloc[:,:-1].values # Specify x
print(x.shape)
y =finally_data["Churn"].values  # Specify y
y.shape
```

```python
from sklearn.decomposition import PCA # use pca to demetion reduction

pca=PCA(n_components=0.9).fit_transform(x)
pca.shape
```

```python
from sklearn.model_selection import KFold  # use k-fold
kf=KFold(n_splits=10)
```

```python
from sklearn.linear_model import LogisticRegression  # train models 
from sklearn.naive_bayes import GaussianNB
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import  precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve


import scikitplot as skplt

i=1
for train_index, test_index in kf.split(pca):


  print("fold : " ,i)
  print("TRAIN:", len(train_index), "TEST:", len(test_index))
  
  x_train, x_test = pca[train_index], pca[test_index]
  
  y_train, y_test = y[train_index], y[test_index]

  lr=LogisticRegression(C=0.001)
  lr.fit(x_train,y_train)
  y_pred_lr=lr.predict(x_test)


  y_pred_proba_lr = lr.predict_proba(x_test)





  conf_matrix=confusion_matrix(y_test,y_pred_lr)
  TP = conf_matrix[1][1]
  TN = conf_matrix[0][0]
  FP = conf_matrix[0][1]
  FN = conf_matrix[1][0]
  specificity=(TN / float(TN + FP))
  precision_lr=precision_score(y_test,y_pred_lr)
  FDR_lr=1-precision_lr
  FOR_lr=TN /float(TN + FN )
  print("***************************************  LogisticRegression  ********************************************")
  plot_roc_curve(lr, x_test, y_test) 
  plt.title(f"ROC curve LogisticRegression fold {i}")
  plt.show()


  skplt.metrics.plot_cumulative_gain(y_test, y_pred_proba_lr)
  plt.show()



  print("LogisticRegression accuracy_score : " ,accuracy_score(y_test,y_pred_lr))
  print("LogisticRegression sensitivity : " ,recall_score(y_test,y_pred_lr))
  print("LogisticRegression specificity : " ,specificity)


  print("LogisticRegression precision_score : " ,precision_lr)
  print("LogisticRegression FDR : " ,FDR_lr)
  print("LogisticRegression FOR : " ,FOR_lr)

  print("LogisticRegression roc_auc_score : " ,roc_auc_score(y_test,y_pred_lr))

















  print("*****************************************  naive_bayes  ********************************************")
  nb=GaussianNB()
  nb.fit(x_train,y_train)
  y_pred_nb=nb.predict(x_test)

  conf_matrix=confusion_matrix(y_test,y_pred_nb)
  TP = conf_matrix[1][1]
  TN = conf_matrix[0][0]
  FP = conf_matrix[0][1]
  FN = conf_matrix[1][0]
  specificity=(TN / float(TN + FP))
  precision_nb=precision_score(y_test,y_pred_nb)
  FDR_nb=1-precision_nb
  FOR_nb=TN /float(TN + FN )




  plot_roc_curve(nb, x_test, y_test,color="r") 
  plt.title(f"ROC curve naive_bayes fold {i}")
  plt.show()
  y_pred_proba_nb = nb.predict_proba(x_test)

  skplt.metrics.plot_cumulative_gain(y_test, y_pred_proba_nb)
  plt.show()


  print("naive_bayes accuracy_score : " ,accuracy_score(y_test,y_pred_nb))
  print("naive_bayes sensitivity : " ,recall_score(y_test,y_pred_nb))
  print("naive_bayes specificity : " ,specificity)

  print("naive_bayes precision_score : " ,precision_nb)
  print("naive_bayes FDR : " ,FDR_nb)
  print("naive_bayes FOR : " ,FOR_nb)


  print("naive_bayes roc_auc_score : " ,roc_auc_score(y_test,y_pred_nb))

  print("===============================")
  
  i+=1
```
